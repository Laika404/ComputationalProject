# Statistical Analysis

As a little extra we quickly did some cursory statistical analysis. Because none of us have any significant statistical background we don't want to draw any big conclusions from this work, but may contain interesting insights.

At a glance it looks that most densities have obvious differences between human-based and centralised models; the means are not close and the standard deviations don't indicate any overlap. With a densities greater than 110 vehicles per kilometre this changes, so we checked if this is likely due to the differences between models and not because of randomness. For this we use $p = 0.05$.

```python
plt.hist(flow_central, bins=10, color="red", alpha=0.5, label="Central control")
plt.hist(flow_agents, bins=10, color="blue", alpha=0.5, label="Human-based")
plt.xlabel("Flow (veh/h)")
plt.title(f"Flow for different behaviours with density = {density:.1f} veh/h")
plt.legend()
plt.show();
plt.hist(speed_central, bins=10, color="red", alpha=0.5, label="Central control")
plt.hist(speed_agents, bins=10, color="blue", alpha=0.5, label="Human-based")
plt.title(f"Mean Speed for different behaviours with density = {density:.1f} veh/h")
plt.xlabel("Mean Speed (m/s)")
plt.legend()
plt.show();
```

```python
plt.boxplot([flow_central, flow_agents], tick_labels=["Central control", "Human-based"])
plt.title(f"Flow for different behaviours with density = {density:.1f} veh/h")
plt.ylabel("Flow (veh/h)")
plt.show();
plt.boxplot([speed_central, speed_agents], tick_labels=["Central control", "Human-based"])
plt.title(f"Mean speed for different behaviours with density = {density:.1f} veh/h")
plt.ylabel("Mean speed (m/s)")
plt.show();
```

With these histograms and boxplots we see that the flows are not at all normally distributed, while the mean speeds are. This makes it easier to analysis on the mean speeds; as test generally assume that the compared values are normally distributed or at least have the same distribution. Now we do a quantitative test of the normality of these distribution, and finally finish with a test that should tell us that the distributions are significantly different.

```python
import scipy
print(scipy.stats.shapiro(speed_central))
print(scipy.stats.shapiro(speed_agents))

print(scipy.stats.ttest_ind(speed_central, speed_agents))
```

The the normality test (Shapiro-Wilk test) indicates that these distributions are indeed normal. Furthermore, because the null hypothesis of the t-test (that the means of the centralised and human-based models' mean speeds are equal) is rejected, we have shown that the results are significantly different.
